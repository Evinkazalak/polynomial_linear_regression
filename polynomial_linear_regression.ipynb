```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("polynomial+regression.csv",sep = ";")
```

```python
y = df.araba_max_hiz.values.reshape(-1,1)
x = df.araba_fiyat.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.ylabel("araba_max_hiz")
plt.xlabel("araba_fiyat")
plt.show()
```

```python
#linear regression =  y = b0 + b1*x
# multiple linear regression   y = b0 + b1*x1 + b2*x2
```

```python
# %% linear regression
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x,y)
```

```python
import numpy as np
y_head = lr.predict(x)
plt.plot(x, y_head, color="red", label="linear")
plt.scatter(x, y, color="blue", label="data points")  # Gerçek veri noktalarının grafiğe eklenmesi
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Tek bir tahmin yapılması (10000 değeri için)
x_pred = np.array([[10000]])  # Tek bir örneği 2B bir diziye dönüştürme
prediction = lr.predict(x_pred)
print("10 milyon TL'lik araba hızı tahmini:", prediction)
```

```python
# polynomial regression =  y = b0 + b1*x +b2*x^2 + b3*x^3 + ... + bn*x^n

from sklearn.preprocessing import PolynomialFeatures
polynomial_regression = PolynomialFeatures(degree = 2)

x_polynomial = polynomial_regression.fit_transform(x)
```

```python
# %% fit
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial,y)
```

```python
y_head2 = linear_regression2.predict(x_polynomial)

plt.plot(x,y_head2,color= "green",label = "poly")
plt.legend()
plt.show()
```
